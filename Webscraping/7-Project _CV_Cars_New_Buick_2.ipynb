## Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
import time
```

```python
# create dataframe (in the beginning it is empty)
df_cars = pd.DataFrame(columns=['New_used', 'Title', 'Mileage', 'Exterior_color', 'Interior_color', 'Drivetrain', 'Fuel_type', 'Transmission', 'Engine', 
                                'Comfort', 'Interior_design', 'Performance', 'Value_for_money', 'Exterior_styling', 'Reliability', 'Overall_score',
                                'Seller_score', 'Price', 'url'])

# Sequence of numbers - 1 to 500 (501 is not included)
for i in range(1,501):
    time.sleep(1)
    
    # variable for pages within website
    website = 'https://www.cars.com/shopping/results/?page='+str(i)+'&page_size=20&list_price_max=&makes[]=buick&maximum_distance=all&models[]=&stock_type=new&zip='
    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'}
    
    # request
    response = requests.get(website, headers=headers)
    
    # create soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # result
    result = soup.find_all('div', {'phx-hook':'VehicleCard'})
    
    # we combine url part no. 1 with url part no. 2 in order to get the absolute url
    url_part_1 = 'https://www.cars.com'
    
    url_part_2 = []

    # Loop through the results
    for item in result:
        # Loop through links
        for link in item.find_all('a', {'class':'vehicle-card-link js-gallery-click-link'}):
            url_part_2.append(link.get('href'))
            
    url_joined = []

    for link_2 in url_part_2:
        url_joined.append(urllib.parse.urljoin(url_part_1, link_2))
        
    for link in url_joined:
        time.sleep(1)
        
        response = requests.get(link, headers=headers)
        
        # create soup object
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # New_used
        try:
            new_used = soup.find('p', {'class':'new-used'}).get_text()
        except:
            new_used = 'n/a'
            
        # Title
        try:
            title = soup.find('h1', {'class':'listing-title'}).get_text()
        except:
            title = 'n/a'
            
        # Mileage
        try:
            mileage = soup.find('div', {'class':'listing-mileage'}).get_text()
        except:
            mileage = 'n/a'
            
        # Exterior_color
        try:
            ext_color = soup.findAll({'dd'})
            a = ext_color[0].get_text()
        except:
            a = 'n/a'
        
        # Interior_color
        try:
            int_color = soup.findAll({'dd'})
            b = int_color[1].get_text()
        except:
            b = 'n/a'
            
        # Drivetrain
        try:
            drivetrain = soup.findAll({'dd'})
            c = drivetrain[2].get_text()
        except:
            c = 'n/a'
        
        # Fuel_type
        try:
            fuel_type = soup.findAll({'dd'})
            d = fuel_type[4].get_text()
        except:
            d = 'n/a'
        
        # Transmission
        try:
            transmission = soup.findAll({'dd'})
            e = transmission[5].get_text()
        except:
            e = 'n/a'
        
        # Engine
        try:
            engine = soup.findAll({'dd'})
            f = engine[6].get_text()
        except:
            f = 'n/a'
        
        # Comfort
        try:
            comfort = soup.findAll('span', {'class':'sds-definition-list__value'})
            g = comfort[0].get_text()
        except:
            g = 'n/a'
        
        # Interior_design
        try:
            int_des = soup.findAll('span', {'class':'sds-definition-list__value'})
            h = int_des[1].get_text()
        except:
            h = 'n/a'
        
        # Performance
        try:
            performance = soup.findAll('span', {'class':'sds-definition-list__value'})
            i = performance[2].get_text()
        except:
            i = 'n/a'
        
        # Value_for_money
        try:
            vfm = soup.findAll('span', {'class':'sds-definition-list__value'})
            j = vfm[3].get_text()
        except:
            j = 'n/a'
        
        # Exterior_styling
        try:
            ext_sty = soup.findAll('span', {'class':'sds-definition-list__value'})
            k = ext_sty[4].get_text()
        except:
            k = 'n/a'
        
        # Reliability
        try:
            reliability = soup.findAll('span', {'class':'sds-definition-list__value'})
            l = reliability[5].get_text()
        except:
            l = 'n/a'
        
        # Overall_score
        try:
            os = soup.findAll('span', {'class':'sds-rating__count'})
            m = os[1].get_text()
        except:
            m = 'n/a'
        
        # Seller_score
        try:
            ss = soup.findAll('span', {'class':'sds-rating__count'})
            n = ss[0].get_text()
        except:
            n = 'n/a'
        
        # Price
        try:
            price = soup.find('span', {'class':'primary-price'}).get_text()
        except:
            price = 'n/a'
            
        # Pandas DataFrame
        df_cars = df_cars.append({'New_used': new_used, 'Title': title, 'Mileage': mileage, 'Exterior_color': a, 'Interior_color': b, 'Drivetrain': c, 'Fuel_type': d,
                                  'Transmission': e, 'Engine': f, 'Comfort': g, 'Interior_design': h, 'Performance': i, 'Value_for_money': j, 'Exterior_styling': k,
                                  'Reliability': l, 'Overall_score': m, 'Seller_score': n, 'Price': price, 'url': link}, ignore_index=True)
```
