### Import the required libraries

```python
import pandas as pd
import os
import xlsxwriter
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
```

### Create variables for input and output location

```python
"""From the input_loc variable we get th files needed.
The output_loc variable will be the storage folder for the appended file."""

input_loc = 'E:/Proyects/Python proyects/WEBSCRAPING/Project_CV_Cars/'

output_loc = 'E:/Proyects/Python proyects/WEBSCRAPING/Project_CV_Cars_Final_File/'
```

### Create variable for list of input files

```python
file_list = os.listdir(input_loc)
file_list
```

### Append the files using a for loop

```python
"""We need to store the data into an excel file, but excel can only support
65000 lines with url`s, so we will store the url`s as string."""

final_df = pd.DataFrame()

for files in file_list:
    if files.endswith('.xlsx'):
        df = pd.read_excel(input_loc + files)
        final_df = final_df.append(df)

writer = pd.ExcelWriter(output_loc + 'final_cars.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})        
#final_df.to_excel(output_loc + 'final_cars.xlsx', index=False, engine='xlsxwriter',options={'strings_to_urls': False})
final_df.to_excel(writer, index=False)
writer.close()
```

### Replacing faulty data

```python
"""We remove the 'mi.' from 'Mileage' so we can transform it to numerical value."""

df_cars['Mileage'] = df_cars['Mileage'].str.replace(r'\D', '')
df_cars
```

```python
"""We remove the '$' from 'Price' so we can transform it to numerical value."""

df_cars['Price'] = df_cars['Price'].str.replace(r'\D', '')
df_cars
```

### Getting the year of the cars from the title

```python
"""We create a new column, 'Year', and using slicing we take the first 4 characters to get the year."""

df_cars['Year'] = df_cars['Title'].str.slice(0,4)
df_cars
```

### Change the position of the 'Year' column

```python
"""To change the position of a specific column, in this case ['Year'].""" 

cols = list(df_cars.columns.values)
df_cars = df_cars[cols[0:2] + [cols[-1]] + cols[2:20]]

df_cars.head(5)
```

### Getting just the model of the car from the 'Title'

```python
"""we create a new column 'Model' and using slicing we take from the fifth character until the end."""

df_cars['Model'] = df_cars['Title'].str.slice(5,99)
df_cars
```

### Change the position of the 'Model' column

```python
"""to change the position of a specific column, in this case 'Model'."""

cols = list(df_cars.columns.values)
df_cars = df_cars[cols[0:3] + [cols[-1]] + cols[3:21]]

df_cars.head(5)
```
